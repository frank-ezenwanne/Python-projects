```python
def take_first_letter(x):
    if pd.notna(x) == False:
        return np.NaN
    return x[0]

def remove_white(x):
    if pd.notna(x) == False:
        return np.NaN
    return x.strip()

df_clean["first_letter_brand"] = df_clean["product"].apply(take_first_letter)
df_clean["first_letter_generic"] = df_clean["GenericName"].apply(take_first_letter)
df_clean["product"] = df_clean["product"].apply(remove_white)
df_clean["GenericName"] = df_clean["GenericName"].apply(remove_white)
```
